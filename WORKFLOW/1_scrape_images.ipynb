```python
import requests
import pandas as pd
```

One of the dataframes from the study gives us the URL's for each image. So we will scrape them and put all of them into one directory from where we will read in the images in later notebooks.

**Figure out URL info**

```python
wikiart = pd.read_csv('../dataset_WikiArt_Emotions/WikiArt-info.tsv', sep='\t')
```

```python
wikiart.info()
```

```python
wikiart.head(3)
```

```python
url_list = wikiart["Image URL"].to_list()
url_list
```

**Scraping**

```python
index = 0 
for url in url_list: 
    response = requests.get(url)
    file = open(f"{wikiart.ID.iloc[index]}.png", "wb")
    file.write(response.content)
    file.close()
    index +=1
#     time.sleep(2)
```
